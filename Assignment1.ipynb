## Importing libraries

```python
import requests
from bs4 import BeautifulSoup
import nltk
import re
from nltk.corpus import stopwords, wordnet
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

## Data

```python
# Step 1: Extract HTML from URL
url = "https://en.wikipedia.org/wiki/Wikipedia"
response = requests.get(url)
html_content = response.text
print(html_content)
```

```python
# Step 2: Extract text from HTML page
soup = BeautifulSoup(html_content, "html.parser")
text_content = ' '.join([p.get_text() for p in soup.find_all(['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6'])])
print(text_content)
```

# Processing on Data

## 1. Cleaning data

```python
# remove square brackets and anything inside them.
text_content = re.sub(r'\[[^\]]*\]', '', text_content)

text_content = re.sub(r'\b[IVXLCDM]+\b', '', text_content)

text_content = re.sub(r'\d+', '', text_content)
print(text_content)
```

```python
# Step 1: Cleaning data
cleaned_text = ''.join(e for e in text_content if e.isalnum() or e.isspace())
print(cleaned_text)
```

## 2. Normalization

```python
# Step 2: Normalization
cleaned_text = cleaned_text.lower()
```

## 3. Tokenization

```python
# Step 3: Tokenization
tokens = word_tokenize(cleaned_text)
tokens
```

## 4. Lemmatization

```python
# Step 4: Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
lemmatized_tokens
```

```python
# Display the original tokens that were tokenized and their corresponding lemmatized forms
for original_token, lemmatized_token in zip(tokens, lemmatized_tokens):
    if original_token != lemmatized_token:
        print(f"Original: {original_token} | Lemmatized: {lemmatized_token}")
```

```python
# Step 4: Lemmatization with specified part of speech
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = []
for word in tokens:
    # Get the part of speech for each word
    pos_tag = nltk.pos_tag([word])[0][1][0].upper()
    # Map the part of speech to WordNet constant
    pos_map = {"N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV, "J": wordnet.ADJ}
    wordnet_pos = pos_map.get(pos_tag, wordnet.NOUN)  # Default to noun if not recognized
    # Lemmatize the word with the specified part of speech
    lemma = lemmatizer.lemmatize(word, pos=wordnet_pos)
    lemmatized_tokens.append(lemma)
```

```python
# Display the original tokens that were tokenized and their corresponding lemmatized forms
for original_token, lemmatized_token in zip(tokens, lemmatized_tokens):
    if original_token != lemmatized_token:
        print(f"Original: {original_token} | Lemmatized: {lemmatized_token}")
```

## 5. Removing stop words

```python
# Step 5: Remove stop words
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in lemmatized_tokens if token not in stop_words]
filtered_tokens
```

## 6. Unique words

```python
# Step 6: Unique words
unique_words = set(filtered_tokens)
unique_words
```

## 7. Removing non-meaningful words

```python
# Step 7: Remove non-meaningful words
meaningful_words = []
for word in unique_words:
    if wordnet.synsets(word):
        meaningful_words.append(word)

meaningful_words
```
